```python
def data_cleaning(data, thresh):
    data = data.dropna(axis=0,thresh = thresh)
    data = data.reset_index(drop=True)
    return data
```

```python
def count_rating(data):
    rating_type = ['AA', 'A', 'BBB', 'BB']
    count_num = [0,0,0,0,0]
    list_a = data['rating'].values.tolist()
    for i in list_a:
        fcnt=0
        for w in range(len(rating_type)):
            if fcnt==0 and rating_type[w] in i:
                count_num[w]=count_num[w]+1
                fcnt=1
            if fcnt==1:
                break
        if fcnt==0:
            count_num[-1]+=1
    return count_num
```

```python
def Y_var(df):
    l = list()
    for e in range(len(df['rating'])):
        if 'AA' in df['rating'].iloc[e]:
            l.append(0)
        elif 'A' in df['rating'].iloc[e]:
            l.append(1)
        elif 'BBB' in df['rating'].iloc[e]:
            l.append(2)
        elif 'BB' in df['rating'].iloc[e]:
            l.append(3)
        else:
            l.append(4)
    return l
```
